```python
# импортируем модули

import string
import re
```

```python
# открываем файл, читаем его, понижаем регистр, убираем пустые строки
with open('city_smells.txt') as f:
    text = f.read()
    text_low = text.lower()
    text_norm = text_low.replace('\n\n', '\n')
text_norm
```

```python
# очищаем текст от пунктуации
text_clean = re.sub('[!»"#$%&\'()*+,—./:;<=>?@[\\]^_`{|}~«]', '', text_norm)
text_clean
```

```python
# в данном дз пока не стала решать проблему точки в десятичной дроби или слэша в паре типа выбросывыхлопы (= выбросы/выхлопы)
error1 = re.findall('.\.\d', text_norm)
# в чистом тексте эти числа выводятся без точки
print(error1)
```

```python
# делим текст на слова, подсчитываем количество слов в чистом тексте
text_words = text_clean.split()
print(text_words)
```

```python
# определить среднюю длину слова в тексте
# определяем количество символов в тексте всего
len_symbols = len(text_clean)
# определяем количество слов в тексте
len_words = len(text_words)
print('Количество символов в тексте:', '\n', len_symbols)
print('Количество слов в тексте:', '\n', len_words)
print('Средняя длина слова в тексте:', '\n', len_symbols/len_words)
```

```python
# определить среднюю длину предложения в тексте
# делим исходный текст на предложения
with open('city_smells.txt') as f:
    text_lines = f.read().splitlines() 
    print('Всего предложений в тексте:', '\n', len(text_lines))
```

```python
# считаем, сколько слов во всех предложениях, и находим среднее количество слов в предложении
# чистим текст
text_lines_split = str(text_lines)
text_lower = text_lines_split.lower()
text_lower_clean = re.sub('[!»"#$%&\'()*+,—./:;<=>?@[\\]^_`{|}~«]', '', text_lower)
text_lines_clean = text_lower_clean.split()
# находим длину всех слов в тексте
len_words_city = len(text_lines_clean)
# делим длину всех слов в тексте на количество предложений
print('Среднее количество слов в предложении:', '\n', len_words_city/len(text_lines))
```

```python
# посчитать, во сколько раз самое длинное предложение длиннее самого короткого (такое же можно сделать со словами)
long_sen = max(text_lines, key = len)
short_sen = min(text_lines, key = len)
len_long_sen = len(long_sen)
len_short_sen = len(short_sen)
res = len(long_sen)/len(short_sen)
print(long_sen, '\n', short_sen)
print('Длина равна:', '\n', len_long_sen, '\n', len_short_sen)
print('Самое длинное предложение длиннее самого короткого в:', res, 'раз')
```

```python
# посчитать, во сколько раз самое длинное слово длиннее самого короткого
long_word = max(text_words, key = len)
short_word = min(text_words, key = len)
len_long_word = len(long_word)
len_short_word = len(short_word)
res1 = len(long_word)/len(short_word)
print(long_word, '\n', short_word)
print('Длина равна:', '\n', len_long_word, '\n', len_short_word)
print('Самое длинное слово длиннее самого короткого в:', res1, 'раз')
```

```python
# (не убирая пунктуацию) - среднее количество пунктуационных знаков в предложении
len_punct = []
with open('city_smells.txt') as f:
    text_raw = f.read()
    for i in text_raw:
        if i in '!»"#$%&\'()*+,—./:;<=>?@[\\]^_`{|}~«':
            len_punct.append(i)
    length_punct = len(len_punct)
    print(len_punct, '\n', 'Всего пунктуационных знаков:', length_punct)
```

```python
print('Среднее количество пунктуационных знаков в предложении:', '\n', length_punct/len(text_lines))
```

```python
# количество уникальных слов и пропорция общему количеству слов в тексте
all_words = len(text_words)
unique = len(set(text_words))
prop = unique/all_words
print(all_words, '\n', unique)
print('Пропорция уникальных слов к общему количеству слов:', '\n', prop)
```

```python
# что-то еще, что Вы сами захотите поисследовать
# найти все слова, которые начинаются на букву "в"
v_words = []
with open('city_smells.txt') as f:
    mytext = f.read()
    for i in mytext:
        if i.startswith('в'):
            v_words.append(i)
    print('Всего слов, начинающихся на букву "в":', '\n', len(v_words))
```
